```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
run_1 = []
run_2 = []
run_3 = []
run_4 = []

pt_1 = (5, -3)
pt_2 = (-3, -2)
pt_3 = (0, 3)
pt_4 = (5, 3)
```

```python
with open("./lab9_data_2/run_1.pkl", "rb") as pickleFile:
        run_1 = pickle.load(pickleFile)
with open("./lab9_data_2/run_2.pkl", "rb") as pickleFile:
        run_2 = pickle.load(pickleFile)
with open("./lab9_data_2/run_3.pkl", "rb") as pickleFile:
        run_3 = pickle.load(pickleFile)
with open("./lab9_data_2/run_4.pkl", "rb") as pickleFile:
        run_4 = pickle.load(pickleFile)
```

```python
# convert distance to num of tiles

run_1 = np.array(run_1) / 333
run_2 = np.array(run_2) / 333
run_3 = np.array(run_3) / 333
run_4 = np.array(run_4) / 333
```

```python
print(run_1, "\n", run_2, "\n", run_3, "\n", run_4)
```

```python
# calculate angle increments

angle = 2 * np.pi / len(run_1)
angle_offset = 2*np.pi - 0.20*np.pi
```

```python
data_1_x = []
data_1_y = []
data_2_x = []
data_2_y = []
data_3_x = []
data_3_y = []
data_4_x = []
data_4_y = []
```

```python
# simualtion with gaussian
for _ in range(10):
    for idx, val in enumerate(run_1):
        data_1_x.append(pt_1[0] - np.cos(angle_offset - angle*idx) * val + np.random.normal(0, 0.2,1))
        data_1_y.append((pt_1[1] - np.sin(angle_offset - angle*idx) * val) + np.random.normal(0, 0.2,1))
    for idx, val in enumerate(run_2):
        data_2_x.append(pt_2[0] - np.cos(angle_offset - angle*idx) * val + np.random.normal(0, 0.2,1))
        data_2_y.append((pt_2[1] - np.sin(angle_offset - angle*idx) * val) + np.random.normal(0, 0.2,1))
    for idx, val in enumerate(run_3):
        data_3_x.append(pt_3[0] - np.cos(angle_offset - angle*idx) * val + np.random.normal(0, 0.2,1))
        data_3_y.append((pt_3[1] - np.sin(angle_offset - angle*idx) * val) + np.random.normal(0, 0.2,1))
    for idx, val in enumerate(run_4):
        data_4_x.append(pt_4[0] - np.cos(angle_offset - angle*idx) * val + np.random.normal(0, 0.2,1))
        data_4_y.append((pt_4[1] - np.sin(angle_offset - angle*idx) * val) + np.random.normal(0, 0.2,1))
```

```python
# points
plt.plot([5, -3, 0, 5], [-3, -2, 3, 3], 'rs')
# region
plt.plot([6, -5, -5, -2, -2, 6, 6], [-4, -4, 1, 1, 4, 4, -4], color='gray')
# obstacles
plt.plot([1, 1, 0, 0], [-4, -2, -2, -4], color='gray')
plt.plot([4, 2, 2, 4, 4], [0, 0, 2, 2, 0], color='gray')
# sensor data
plt.plot(data_1_x, data_1_y, '.',color='brown', label='run 1')
plt.plot(data_2_x, data_2_y, '.',color='blue', label='run 2')
plt.plot(data_3_x, data_3_y, '.',color='green', label='run 3')
plt.plot(data_4_x, data_4_y, '.',color='orange', label='run 4')
plt.legend(loc="upper left")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
